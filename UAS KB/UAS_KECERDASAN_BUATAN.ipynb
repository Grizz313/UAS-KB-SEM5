# **CLUSTERING**

##### Clustering adalah metode pembelajaran tanpa pengawasan yang digunakan untuk menemukan kelompok-kelompok baru dari data sehingga jarak intra-kluster diminimalkan dan jarak antar-kluster dimaksimalkan. Ini tidak memerlukan pengetahuan apa pun tentang data, mirip dengan algoritma klasifikasi. Algoritma clustering memberikan (atau memprediksi) sebuah angka untuk setiap titik data, menunjukkan ke kluster mana titik tertentu tersebut termasuk. Ada dua jenis clustering yang berbeda:

* PARTITIONAL
* HIERARCHICAL

##### Hierarchical clustering hanya memerlukan ukuran kesamaan, sedangkan partitional clustering memerlukan asumsi yang lebih kuat seperti jumlah kluster dan pusat awal. Hierarchical clustering menghasilkan pembagian kluster yang lebih bermakna dan subjektif, tetapi partitional clustering menghasilkan tepat k kluster.

##### Algoritma clustering paling umum adalah: K-Means, Hierarchical, Density-based (DBSCAN).

# **K-MEANS CLUSTERING**

##### K-Means clustering adalah salah satu algoritma pengelompokan paling sederhana dan umum digunakan. Ini berusaha menemukan pusat-pusat kelompok yang mewakili wilayah-wilayah tertentu dari data. Algoritma ini bergantian antara dua langkah: menetapkan setiap titik data ke pusat kelompok terdekat, dan kemudian menetapkan setiap pusat kelompok sebagai rata-rata dari titik-titik data yang ditetapkan kepadanya. Algoritma selesai ketika penugasan instansi ke kelompok tidak lagi berubah.

##### Ini adalah pendekatan partisional lengkap, memerlukan metrik (cara mengukur jarak) dan data harus dinormalisasi. Cara paling umum untuk mengukur jarak adalah dengan jumlah kesalahan kuadrat:

# **SSE = ∑i=0n d(xi, cj)2**

##### K-Means memiliki keuntungan bahwa ini cukup cepat, karena yang sebenarnya kita lakukan hanyalah menghitung jarak antara titik-titik dan pusat kelompok. Oleh karena itu, memiliki kompleksitas linear O(n).

* **ELBOW METHOD**

#####Mungkin untuk menemukan nilai terbaik dari K pada grafik SSE dengan variasi jumlah K, dari grafik tersebut Anda memilih nilai K di mana ada peningkatan kemiringan yang lebih tinggi, dalam kasus kami K=2.

```python
wcss = []
for i in range(1,8):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=7,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,8),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

* **K-MEANS**

```python
def K_Means(X, n):
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    model = KMeans(n)
    model.fit(X)
    clust_labels = model.predict(X)
    cent = model.cluster_centers_
    return (clust_labels, cent)
```

```python
def Plot3dClustering(n, X, type_c):
    data = []
    clusters = []
    colors = ['rgb(228,26,28)','rgb(55,126,184)','rgb(77,175,74)']

    for i in range(n):
        name = i
        color = colors[i]
        x = X[ X[type_c] == i ]['Total']
        y = X[ X[type_c] == i ]['Population']
        z = X[ X[type_c] == i ]['Surface']

        trace = dict(
            name = name,
            x = x, y = y, z = z,
            type = "scatter3d",
            mode = 'markers',
            marker = dict( size=4, color=color, line=dict(width=0) ) )
        data.append( trace )

        cluster = dict(
            color = color,
            opacity = 0.1,
            type = "mesh3d",
            alphahull = 7,
            name = "y",
            x = x, y = y, z = z )
        data.append( cluster )

    layout = dict(
        width=800,
        height=550,
        autosize=False,
        title='3D Clustering Plot',
        scene=dict(
            xaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                title='Total Production',
                backgroundcolor='rgb(230, 230,230)'
            ),
            yaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                title='Population',
                backgroundcolor='rgb(230, 230,230)'
            ),
            zaxis=dict(
                gridcolor='rgb(255, 255, 255)',
                zerolinecolor='rgb(255, 255, 255)',
                showbackground=True,
                title='Surface Area',
                backgroundcolor='rgb(230, 230,230)'
            ),
            aspectratio = dict( x=1, y=1, z=0.7 ),
            aspectmode = 'manual'
        ),
    )

    fig = dict(data=data, layout=layout)
    iplot(fig, filename='total_surface_population_plot', validate=False)
```
